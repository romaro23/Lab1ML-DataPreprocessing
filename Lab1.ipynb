```python
import numpy
import matplotlib
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from pandas.core.interchange.dataframe_protocol import DataFrame

pd.set_option('display.max_columns', None)
#4
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'income'
]
#2
df = pd.read_csv("adult/adult.data", names=column_names, skipinitialspace=True)
```

```python
#3
print("Кількість рядків та стовпців:\n", df.shape)
print("Типи стовпців:\n", df.dtypes)
print("Перші 5 рядків:\n")
df.head()
```

```python
#5
print("Стовпець age:\n", df['age'])
print("Стовпці education та fnlwgt:\n", df[['education', 'fnlwgt']])
print("Перший рядок:\n", df.iloc[0])
print("Рядки з п'ятого по дев'ятий:\n", df.iloc[5:10])
print("Перший та другий стовпець для рядків з першого по четвертий:\n", df.iloc[0:5, [0, 1]])
df['education'] = df['education'].str.strip()
filtered_df = df[df['education'] == 'Bachelors']
print("Рядки де education = Bachelors:\n")
filtered_df
```

```python
df['sex'] = df['sex'].str.strip()
df['income'] = df['income'].str.strip()
print(df[(df['income'] == '>50K') & (df['sex'] == 'Female')].loc[:, ['fnlwgt', 'age']])
```

```python
#6
quantitative_columns = ['age', 'fnlwgt', 'capital-gain', 'capital-loss', 'hours-per-week']
quantitative_df = df[quantitative_columns]
stats = quantitative_df.describe()
print(stats)
median = quantitative_df.median()
print("median\n ", median)
unique_counts = quantitative_df.nunique()
print("Кількість унікальних значень у кількісних ознаках:\n", unique_counts)
print("Масштаби ознак:\n", quantitative_df.max() - quantitative_df.min())
```
